```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://webscraper.io/test-sites/e-commerce/allinone/computers/tablets"
r = requests.get(url)
print(r)
```

```python
print(r.__attrs__)
```

```python
soup = BeautifulSoup(r.text, "html.parser")
print(soup)
```

```python
tag = soup.comments
# atb = tag.attrs
# print(atb["role"])
print(tag)
```

```python
print(soup.h4)
```

```python
# find()function

title = soup.find("a", {"class": "title"})
print(title.string)

des = soup.find("p", {"class": "description card-text"})
print(des.string)

price = soup.find("span", {"itemprop": "price"})
print(price.string)
```

```python
# find_all()
# prices = soup.find_all("span", {"itemprop": "price"})
# # print(len(prices))
# for i in prices:
#   print(i.text)

# titles = soup.find_all("a", {"class": "title"})
# # print(titles)
# for i in titles:
#   print(i.text)

dess = soup.find_all("p", {"class": "description card-text"})
# print(dess)
for i in dess:
  print(i.text)
```

```python
# regular expressions with bs4
import re

data = soup.find_all(string=re.compile('7"'))
# print(data)
for i in data:
  print(i)
```

```python
# BS4 with Pandas; Extract name, description and price

import pandas as pd

# Product name
product_name = []
titles = soup.find_all("a", {"class": "title"})
for i in titles:
  product_name.append(i.text)
print(product_name)

# Description
product_description = []
desc = soup.find_all("p", {"class": "description card-text"})
for i in desc:
  product_description.append(i.text)
print(product_description)

# Price
price = []
prices = soup.find_all("span", {"itemprop": "price"})
for i in prices:
  price.append(i.text)
print(price)

# rating = []
# ratings = soup.find_all()

df = pd.DataFrame({"Product Name":product_name, "Description": product_description, "Price":price})
df

# df.to_csv("products.csv")
```

```python
#Nested HTML Tags
# divs = soup.find_all("div", {"class":"col-md-4 col-xl-4 col-lg-4"})
# print(len(divs))

divs = soup.find_all("div", {"class":"col-md-4 col-xl-4 col-lg-4"})[3]
print((divs))
```

```python
url2 = "https://webscraper.io/test-sites/e-commerce/allinone/computers"
req = requests.get(url2)
```

```python
sp = BeautifulSoup(req.text, "html.parser")
# print(sp)
```

```python
names = sp.find_all("span",{"itemprop":"name"})
for i in names:
  print(i.text)
```

```python
menu = sp.find("ul", {"class":"nav flex-column"}, {"id":"side-menu"})
print(menu)
```

```python
category = []
cat = menu.find_all("li", {"class":"nav-item"})
for i in cat:
  category.append(i.text)
print(category)
```
