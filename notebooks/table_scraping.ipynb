```python
#Libraries
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
# url = "https://ticker.finology.in/"
url = "https://ticker.finology.in/market/52-week-high"
r = requests.get(url)
# print(r)
```

```python
soup = BeautifulSoup(r.text, "html.parser")

table = soup.find("table", {"data-toggle":"table"})
print(table)
```

```python
#Extract the table headers

headers = []
tbl_headers = table.find_all("th")
for i in tbl_headers:
  headers.append(i.text)
# print(headers)

df = pd.DataFrame(columns=headers)
df
```

```python
rows = table.find_all("tr")
# print(rows)
#start fro the first index to exclude the headers
for i in rows[1:]:
   data = i.find_all("td")
   row = [tr.text for tr in data]
  #  print(row)
   l = len(df)
   df.loc[l] = row

print(df)
```
